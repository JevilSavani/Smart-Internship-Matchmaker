```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def scrape_internshala_pages(start=1, end=199):
    all_data = []

    for a in range(start, end):
        url = f'https://internshala.com/internships/page-{a}/'
        webpage = requests.get(url).text
        soup = BeautifulSoup(webpage, 'lxml')

        company_blocks = soup.find_all('div', class_="internship_meta duration_meta")

        for i in company_blocks:
            name = i.find("p").text.strip()
            stipend = i.find("span", class_="stipend").text.removesuffix("/month")
            location = i.find("a").text

            badge = i.find("div", class_="actively-hiring-badge")
            hiring = badge.text if badge else "Not actively hiring"

            skills = [div.text.strip() for div in i.find_all("div", class_="job_skill")]

            all_data.append({
                "Name": name,
                "location": location,
                "stipend": stipend,
                "skills": skills,
                "hiring": hiring
            })

    return pd.DataFrame(all_data)
```
